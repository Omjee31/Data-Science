```python
# =========================================
# Linear Algebra for Machine Learning (NumPy Notebook)
# =========================================

import numpy as np

# =========================================
# 1. VECTORS & BASIC OPERATIONS
# =========================================
print("\n=== 1. VECTORS ===")
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])

print("Addition:", v1 + v2)
print("Scalar Multiplication:", 3 * v1)
print("Dot Product:", np.dot(v1, v2))
print("Norm:", np.linalg.norm(v1))


# =========================================
# 2. MATRICES
# =========================================
print("\n=== 2. MATRICES ===")
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print("Matrix Addition:\n", A + B)
print("Matrix Multiplication:\n", np.matmul(A, B))
print("Transpose:\n", A.T)


# =========================================
# 3. LINEAR INDEPENDENCE & RANK
# =========================================
print("\n=== 3. RANK & LINEAR INDEPENDENCE ===")
M = np.array([[1, 2, 3], [2, 4, 6], [1, 1, 1]])
rank = np.linalg.matrix_rank(M)
print("Rank:", rank)


# =========================================
# 4. SPAN & BASIS
# =========================================
print("\n=== 4. SPAN & BASIS ===")
vectors = np.array([[1, 0], [0, 1]])
coeff = np.array([2, 3])
print("Span:", np.dot(coeff, vectors))


# =========================================
# 5. COLUMN SPACE & ROW SPACE
# =========================================
print("\n=== 5. COLUMN & ROW SPACE ===")
A = np.array([[1, 2], [3, 4], [5, 6]])

print("Column Space:")
for i in range(A.shape[1]):
    print(A[:, i])

print("Row Space:")
for i in range(A.shape[0]):
    print(A[i, :])


# =========================================
# 6. RREF (ROW REDUCED ECHELON FORM)
# =========================================
print("\n=== 6. RREF ===")

def rref(matrix):
    A = matrix.astype(float)
    rows, cols = A.shape
    r = 0

    for c in range(cols):
        if r >= rows:
            break

        pivot = np.argmax(abs(A[r:, c])) + r
        if A[pivot, c] == 0:
            continue

        A[[r, pivot]] = A[[pivot, r]]
        A[r] = A[r] / A[r, c]

        for i in range(rows):
            if i != r:
                A[i] -= A[i, c] * A[r]

        r += 1

    return A

R = np.array([[1, 2, 3], [2, 4, 6], [1, 1, 1]])
print(rref(R))


# =========================================
# 7. LINEAR TRANSFORMATION
# =========================================
print("\n=== 7. LINEAR TRANSFORMATION ===")
T = np.array([[2, 0], [0, 3]])
v = np.array([1, 2])

print("Transformed Vector:", np.dot(T, v))


# =========================================
# 8. DETERMINANT & INVERSE
# =========================================
print("\n=== 8. DETERMINANT & INVERSE ===")
A = np.array([[1, 2], [3, 4]])

print("Determinant:", np.linalg.det(A))
print("Inverse:\n", np.linalg.inv(A))


# =========================================
# 9. EIGENVALUES & EIGENVECTORS
# =========================================
print("\n=== 9. EIGENVALUES & EIGENVECTORS ===")
A = np.array([[2, 1], [1, 2]])

values, vectors = np.linalg.eig(A)
print("Eigenvalues:", values)
print("Eigenvectors:\n", vectors)


# =========================================
# 10. PROJECTION
# =========================================
print("\n=== 10. PROJECTION ===")
a = np.array([3, 4])
b = np.array([1, 0])

proj = (np.dot(a, b) / np.dot(b, b)) * b
print("Projection:", proj)


# =========================================
# 11. SVD (VERY IMPORTANT FOR ML)
# =========================================
print("\n=== 11. SVD ===")
A = np.array([[1, 2], [3, 4]])

U, S, Vt = np.linalg.svd(A)
print("U:\n", U)
print("S:", S)
print("Vt:\n", Vt)


# =========================================
# 12. PCA (MINI IMPLEMENTATION)
# =========================================
print("\n=== 12. PCA ===")

X = np.array([[2.5, 2.4],
              [0.5, 0.7],
              [2.2, 2.9],
              [1.9, 2.2],
              [3.1, 3.0],
              [2.3, 2.7],
              [2, 1.6],
              [1, 1.1],
              [1.5, 1.6],
              [1.1, 0.9]])

# Step 1: Mean Centering
X_meaned = X - np.mean(X, axis=0)

# Step 2: Covariance Matrix
cov_matrix = np.cov(X_meaned, rowvar=False)

# Step 3: Eigen Decomposition
eigen_values, eigen_vectors = np.linalg.eig(cov_matrix)

# Step 4: Sort eigenvalues
sorted_index = np.argsort(eigen_values)[::-1]
sorted_vectors = eigen_vectors[:, sorted_index]

# Step 5: Projection
X_reduced = np.dot(X_meaned, sorted_vectors[:, :1])

print("Reduced Data:\n", X_reduced)


# =========================================
# END
# =========================================
print("\nAll Linear Algebra operations completed successfully!")
```
