```python
# Matrix multiply
#A neural layer has weights W of shape (8, 4) and receives input x of shape (4,). 
# Compute the output manually using @ operator. What is the output shape? What does each output value represent?
weight = np.random.randint(0,10,size=(8,4))
input = np.random.randint(0,5,size= (4,1))
output = np.dot(weight,input)
print("Orignal Weight : \n",weight)
print("Shape Of Output ",output.shape)
print("Ouput :  \n",output)


# You have a batch of 16 samples, each with 6 features. 
# The weight matrix has 10 neurons. Write the full forward pass including a bias vector. Print all shapes.

# 1. Input (batch of 16 samples, each with 6 features)
X = np.random.randn(16, 6)

# 2. Weights (6 input features → 10 neurons)
W = np.random.randn(6, 10)

# 3. Bias (1 bias per neuron)
b = np.random.randn(1, 10)

# ---- Forward Pass ----
Z = np.dot(X, W) + b

# ---- Print Shapes ----
print("X shape:", X.shape)
print("W shape:", W.shape)
print("b shape:", b.shape)
print("Z shape:", Z.shape)
print(Z)
```
