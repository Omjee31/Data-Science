# 📊MatplotLib Scatter Plot Parameters (Important Notes)

- **x** → Values for X-axis (horizontal axis)  
- **y** → Values for Y-axis (vertical axis)  

- **c** → Used to represent a third variable (color encoding)  
  👉 Assigns different colors based on values  

- **cmap** → Colormap used to define color scheme  
  👉 Examples: `"viridis"`, `"plasma"`, `"coolwarm"`  

---

## 🔹 Additional Important Parameters

- **s** → Size of points  
  👉 Can be constant or based on a column  

- **alpha** → Transparency of points  
  👉 Range: 0 (transparent) to 1 (solid)  

- **marker** → Shape of points  
  👉 Examples: `'o'`, `'x'`, `'^'`, `'*'`  

- **edgecolors** → Border color of points  

- **linewidths** → Thickness of point border  

---

## 🔹 Labeling & Styling

- **label** → Name used in legend  

- **plt.xlabel()** → Label for X-axis  
- **plt.ylabel()** → Label for Y-axis  
- **plt.title()** → Title of plot  

- **plt.legend()** → Show legend  

---

## 🔹 Grid & Layout

- **plt.grid()** → Show grid lines  

- **plt.figure(figsize=(w,h))** → Set figure size  

---

## 🔥 Simple Understanding

👉 **x, y** → position  
👉 **c, cmap** → color  
👉 **s** → size  
👉 **alpha** → transparency  
👉 **marker** → shape  

---

## ⚡ Final One-Line Revision

👉 **Scatter plot = position (x,y) + color (c) + size (s) + style**

# 📊 Seaborn Scatter Plot (Important Notes)

- **data** → Dataset used for plotting  

- **x** → Column name for X-axis (horizontal axis)  
- **y** → Column name for Y-axis (vertical axis)  

- **hue** → Used to represent categories with different colors  
  👉 Example: species (setosa, versicolor, virginica)  

- **palette** → Color theme for hue  
  👉 Examples: `"deep"`, `"muted"`, `"pastel`, `"rainbow"`  

---

## 🔹 Additional Important Parameters

- **size** → Size of points based on a variable  

- **sizes** → Control range of sizes  

- **style** → Different marker styles based on category  

- **markers** → Custom marker shapes  

- **alpha** → Transparency of points  

---

## 🔹 Labeling & Styling

- **plt.xlabel()** → Label for X-axis  
- **plt.ylabel()** → Label for Y-axis  
- **plt.title()** → Title of plot  

- **plt.legend()** → Show legend (auto in seaborn)  

---

## 🔹 Grid & Theme

- **sns.set_style()** → Set background style  
  👉 Examples: `"white"`, `"dark"`, `"ticks"`, `"whitegrid"`  

- **sns.set_palette()** → Set default color palette  

---

## 🔹 Example Code

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("iris")

sns.scatterplot(
    data=df,
    x="petal_length",
    y="sepal_length",
    hue="species",
    palette="rainbow"
)

plt.title("Seaborn Scatter Plot")
plt.show()
