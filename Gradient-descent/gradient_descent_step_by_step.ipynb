```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y = make_regression(n_samples=4, n_features=1,n_informative=1,n_targets=1,noise = 80,random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
# Let's apply OLS
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color = 'r')
```

```python
# Lets apply gradient descent assuming slope is constant m = 78.35
# and lets assume the starting value for intercept b = 0
y_pred = ((78.35*X)+100).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color = 'r',label = 'OLS')
plt.plot(X,y_pred,color = 'g',label = 'OLS')
plt.legend()
plt.show()
```

```python
m = 78.35
b = 100
loss_slope = -2*np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
#lets take learning rate = 0.1
lr = 0.1
step_size = loss_slope*lr
step_size
```

```python
#Calculating new intercept
b = b-step_size
b
```

```python
y_pred1 = ((78.35*X)+b).reshape(4)
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color = 'r',label = 'OLS')
plt.plot(X,y_pred,color = 'g',label = 'b = 0')
plt.plot(X,y_pred1,color = 'y',label = 'b = {}'.format(b.round(2)))
plt.legend()
plt.show()
```

```python
#iteration 2 
loss_slope = -2*np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b-step_size
b
```

```python
y_pred2 = ((78.35*X)+b).reshape(4)
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color = 'r',label = 'OLS')
plt.plot(X,y_pred,color = 'g',label = 'b = 0')
plt.plot(X,y_pred1,color = 'y',label = 'b = {}'.format(b.round(2)))
plt.plot(X,y_pred2,color = 'b',label = 'b = {}'.format(b.round(2)))
plt.legend()
plt.show()
```

```python
b = -100
m = 78.35
lr = 0.01
epochs = 100
for i in range (epochs):
    loss_slope = -2 * np.sum(y-m*X.ravel()-b)
    b = b-(lr*loss_slope)

    y_pred = m*X+b
    plt.plot(X,y_pred)

plt.scatter(X,y)
```
