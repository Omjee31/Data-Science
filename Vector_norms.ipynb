## L1 (manhatt)
    it is use to calculate legth(weight) of vector 
    it is calculated by sum of all vector elements

```python
# vector L1 norm
from numpy import array
from numpy.linalg import norm
# define vector
a = array([1, 2, 3,0,0])
print(a)
# calculate norm
l1 = norm(a,1)  # here 1 is subscript   (sum  of all elements)
print(l1)
```

## Vector L2 Norm (Euclidian norm/distance)
    The length of a vector can be calculated using the L2 norm,
    where the 2 is a superscript of the L. 
    The notation for the L2 norm of a vector is ||v||2 where 2 is a subscript.


```python
# vector L2 norm
from numpy import array
from numpy.linalg import norm
# define vector
a = array([1, 2, 3])
print(a)
# calculate norm
l2 = norm(a) # Here we can also use 2 as subscript ,(it is sqrt of l1)
print(l2)
```

## Vector norm max

```python
# vector max norm
from math import inf
from numpy import array
from numpy.linalg import norm
# define vector
a = array([1, 2, 3,19])
print(a)
# calculate norm
maxnorm = norm(a,inf)
print(maxnorm)  # it give max number from vector
```
