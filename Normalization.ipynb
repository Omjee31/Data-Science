## Normalization

Normalization is a technique often applied as part of **data preparation for machine learning**.  
The goal of normalization is to change the values of numeric columns in a dataset to a **common scale**,
without distorting differences in the ranges of values or losing important information.

## Normalization

Normalization is a data preprocessing technique used in **Machine Learning** to scale numeric features to a common range so that all features contribute equally to the model ( **Mostly use in image (cnn)**)

---

## 1. Min-Max Normalization

This method rescales the data to a fixed range, usually **[0,1]**.

Formula:

\[
X_{norm} = \frac{X - X_{min}}{X_{max} - X_{min}}
\]

Where:
- \(X\) = original value  
- \(X_{min}\) = minimum value of the feature  
- \(X_{max}\) = maximum value of the feature  

---

## 2. Z-Score Normalization (Standardization)

This method transforms the data so that it has **mean = 0** and **standard deviation = 1**.

Formula:

\[
Z = \frac{X - \mu}{\sigma}
\]

Where:
- \(X\) = original value  
- \(\mu\) = mean of the feature  
- \(\sigma\) = standard deviation  

---

## 3. L1 Normalization

In L1 normalization, values are scaled so that the **sum of absolute values equals 1**.

Formula:

\[
X_{norm} = \frac{X}{\sum |X|}
\]

---

## 4. L2 Normalization

In L2 normalization, values are scaled based on the **square root of the sum of squares**.

Formula:

\[
X_{norm} = \frac{X}{\sqrt{\sum X^2}}
\]
